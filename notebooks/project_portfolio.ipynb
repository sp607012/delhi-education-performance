# 2. Compare Exam Levels: Performance Change

```python
import matplotlib.pyplot as plt
import pandas as pd

'main code'

df['quality_index_change'] = df['class_xii_quality_index'] - df['class_x_quality_index']

# Preview for all districts
print(df[['district', 'class_x_quality_index', 'class_xii_quality_index', 'quality_index_change']])


df_sorted = df.sort_values('quality_index_change')
y_pos = range(len(df_sorted))

fig, ax = plt.subplots(figsize=(12, 10))

colors = ['green' if x > 0 else 'red' for x in df_sorted['quality_index_change']]
bars = ax.barh(y_pos, df_sorted['quality_index_change'], color=colors, edgecolor='black')

ax.set_yticks(y_pos)
ax.set_yticklabels(df_sorted['district'])
ax.set_xlabel('Quality Index Change (Class XII - Class X)')
ax.set_title('Quality Index Change per District (Class XII vs Class X)')
ax.grid(axis='x', linestyle='--', alpha=0.7)

# Add value labels at the end of each bar
for bar, value in zip(bars, df_sorted['quality_index_change']):
    ax.text(value, bar.get_y() + bar.get_height()/2, 
            f'{value:.2f}', 
            va='center', ha='left' if value > 0 else 'right', 
            color='black', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

```

# 5. Detect Districts with High Improvements or Declines in Quality from Class X to Class XII

```python
import matplotlib.pyplot as plt
import pandas as pd

'main code'

df['quality_index_change'] = df['class_xii_quality_index'] - df['class_x_quality_index']

# Top improvements (largest positive change)
top_improvements = df.nlargest(5, 'quality_index_change')[['district', 'quality_index_change']]

# Top declines (largest negative change)
top_declines = df.nsmallest(5, 'quality_index_change')[['district', 'quality_index_change']]

print("Districts with highest improvements in quality index:")
print(top_improvements)
print("Districts with highest declines in quality index:")
print(top_declines)

'visulization'
# Calculate quality index change if not already done
df['quality_index_change'] = df['class_xii_quality_index'] - df['class_x_quality_index']

# Top improvements (positive changes)
top_improvements = df.nlargest(5, 'quality_index_change')[['district', 'quality_index_change']]
top_improvements['Change Type'] = 'Improvement'

# Top declines (negative changes)
top_declines = df.nsmallest(5, 'quality_index_change')[['district', 'quality_index_change']]
top_declines['Change Type'] = 'Decline'

# Combine for plotting
quality_change_df = pd.concat([top_improvements, top_declines])

# Plotting
plt.figure(figsize=(12,7))
colors = {'Improvement':'green', 'Decline':'red'}
bars = plt.bar(quality_change_df['district'], quality_change_df['quality_index_change'],
       color=quality_change_df['Change Type'].map(colors))

plt.axhline(0, color='black', linewidth=0.8)
plt.title('Top 5 Districts with Highest Improvements and Declines in Quality Index (Class XII vs Class X)')
plt.xlabel('District')
plt.ylabel('Quality Index Change (Class XII - Class X)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add value labels above bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),  # 5 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

plt.show()
```

# What is a Quality Index?
Imagine you're checking how good the teaching or learning is in schools — not just whether students pass, but how well they are being taught and how effectively they are learning.

This is where the Quality Index comes in. It’s like a score or rating that tells us:

“How good is the quality of education in this class or district?”

📘 Simple Breakdown:
It’s not just about pass percentage.
It looks at how strong the performance is — like how many students got high marks, consistency, etc.
It’s often calculated using multiple factors: like attendance, learning levels, assessments, and performance distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

'MAIN CODE'

# 1️⃣ Student–School Ratios
df['students_per_primary_school'] = np.where(
    df['no_of_schools_primary'] > 0,
    df['no_of_students_primary'] / df['no_of_schools_primary'],
    np.nan  # NaN where no schools
)

df['students_per_middle_school'] = np.where(
    df['no_of_schools_middle'] > 0,
    df['no_of_students_middle'] / df['no_of_schools_middle'],
    np.nan
)

# 2️⃣ Flag extreme load cases
high_load_primary = df[df['students_per_primary_school'] > 5000]
high_load_middle = df[df['students_per_middle_school'] > 20000]

# 3️⃣ Districts with 0 schools but students
zero_primary_gap = df[(df['no_of_schools_primary'] == 0) & (df['no_of_students_primary'] > 0)]
zero_middle_gap = df[(df['no_of_schools_middle'] == 0) & (df['no_of_students_middle'] > 0)]





# Assume df is already loaded and matches your table

# Handle division by zero for school counts
df['students_per_primary_school'] = df.apply(
    lambda row: row['no_of_students_primary']/row['no_of_schools_primary'] if row['no_of_schools_primary']>0 else np.nan, axis=1)
df['students_per_middle_school'] = df.apply(
    lambda row: row['no_of_students_middle']/row['no_of_schools_middle'] if row['no_of_schools_middle']>0 else np.nan, axis=1)

# Prepare data
index = np.arange(len(df))
bar_width = 0.35

plt.figure(figsize=(13,6))
bars1 = plt.bar(index, df['students_per_primary_school'], bar_width, label='Primary', color='skyblue')
bars2 = plt.bar(index + bar_width, df['students_per_middle_school'], bar_width, label='Middle', color='orange')

# Add value labels
for bar in bars1:
    val = bar.get_height()
    if not np.isnan(val):
        plt.text(bar.get_x() + bar.get_width()/2, val + 200, f'{val:.0f}', ha='center', va='bottom', fontsize=9)
for bar in bars2:
    val = bar.get_height()
    if not np.isnan(val):
        plt.text(bar.get_x() + bar.get_width()/2, val + 200, f'{val:.0f}', ha='center', va='bottom', fontsize=9)

plt.xlabel('District', fontsize=12)
plt.ylabel('Students per School', fontsize=12)
plt.title('Student Load per School by Level', fontsize=14)
plt.xticks(index + bar_width/2, df['district'], rotation=75)
plt.legend()
plt.tight_layout()
plt.show()

# 5️⃣ Print quick insights
print("⚠ Districts with high load in Primary (>5000 students per school):")
print(high_load_primary[['district','students_per_primary_school']])

print("\n⚠ Districts with high load in Middle (>20000 students per school):")
print(high_load_middle[['district','students_per_middle_school']].round(1))

print("\n🚨 Districts with students but ZERO primary schools:")
print(zero_primary_gap[['district','no_of_students_primary']])

print("\n🚨 Districts with students but ZERO middle schools:")
print(zero_middle_gap[['district','no_of_students_middle']])
```
